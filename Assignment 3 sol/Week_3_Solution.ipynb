```python
import pandas as pd

# Replace 'filename.csv' with your actual file name
df = pd.read_csv('student-por.csv')

# Show the first few rows of the dataset
df.head()
```

```python
# Get basic info about the data
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df['G3'], bins=15, kde=True, color='skyblue')
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.countplot(x='sex', data=df, palette='Set2')
plt.title("Count of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='sex', y='G3', data=df, palette='Pastel1')
plt.title("Final Grade (G3) by Gender")
plt.xlabel("Gender")
plt.ylabel("Final Grade")
plt.show()
```

```python
sns.boxplot(x='studytime', y='G3', data=df, palette='coolwarm')
plt.title("Final Grades by Study Time")
plt.xlabel("Study Time (1 = <2h, 4 = 10+ h)")
plt.ylabel("Final Grade")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='absences', y='G3', data=df, hue='sex')
plt.title("Absences vs Final Grade")
plt.xlabel("Absences")
plt.ylabel("Final Grade")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
numeric_cols = df.select_dtypes(include='number')
sns.heatmap(numeric_cols.corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Matrix of Numeric Variables")
plt.show()
```
